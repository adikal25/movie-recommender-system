```python
import ast
ast.literal_eval('[{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]')
```

```python
#[{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]'
def convert(obj):
    genre_list=[]
    for i in ast.literal_eval(obj):
        genre_list.append(i['name'])
    return genre_list
 


        
    
```

```python
def convert_cast(obj):
    cast=[]
    count=0
    for i in ast.literal_eval(obj):
        cast.append(i['name'])
        
    return cast[0:3]
        
    
        
        
```

```python
#we use NLP(mixbag of combining all the tags of the movies in our dataset and now after creating the big string we cross check the frequency of each word of the big string with the  
# tags of the each move) we dont consider stop words and consider the vectorization
#5000X5000 vector
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
#applying stemming for the similar words (converts the loving to love  )
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies_df['tags']=movies_df['Tags'].apply(stem)
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
vector=cv.fit_transform(movies_df['Tags']).toarray()
```

```python
vector
#now we have 4806 movies with the 5000 most occuring words vectorized
```

```python
cv.get_feature_names_out()
```

```python
#we now use cosine distance that is calculate 
# the dist btw the movie vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector).shape
#calculating dist of a single movie with all the 
# other movies present in the vector
```

```python
similarity=cosine_similarity(vector)
print(similarity)
```

```python
def recommend(movie):
    movie_index=movies_df[movies_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    
    
    for i in movies_list:
        print(movies_df.iloc[i[0]].title)
        
```
